```python
import bs4
import requests
url = "https://www.nosdeputes.fr/deputes"
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, "html.parser")
all_parts = soup.find_all("div", {"class":"list_table"})
for each_div in all_parts:
    find_links = each_div.find_all("a")
    for link in find_links:
        url_2 = "https://www.nosdeputes.fr"+link.get("href")
        response_2 = requests.get(url_2)
        soup_2 = bs4.BeautifulSoup(response_2.text, "html.parser")
        contact_parts = soup_2.find_all("a")
        for a in contact_parts:
            if a.text.endswith('@assemblee-nationale.fr'):
                print(a.text)
```
